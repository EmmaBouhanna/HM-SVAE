```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("dark")
sns.set_palette("husl")
```

```python
samples = 100000
```

```python
fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(20,4))
for i, dim in enumerate([2, 10, 20, 100]):
  res = 0 + np.random.randn(samples, dim)
  res_rescaled = res*np.linalg.norm(res, axis=1)[:, np.newaxis]/(np.sqrt(dim)*np.linalg.norm(res[:,:2], axis=1)[:, np.newaxis])
  ax[i].hist2d(res_rescaled[:,0], res_rescaled[:,1], bins=100, cmap="viridis")
  ax[i].set_ylim((-1.3, 1.2))
  ax[i].set_xlim((-1.3, 1.3))
  ax[i].axis("off")
  ax[i].set_title(f"Dimension {dim}")
  plt.savefig("gaussian_to_hypershere.png", dpi=300)
plt.show()
```
